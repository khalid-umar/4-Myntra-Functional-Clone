```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


# Function to convert DMS to Decimal Degrees
def dms_to_decimal(degree, minute, second):
    """Convert DMS to Decimal Degrees"""
    return degree + (minute / 60) + (second / 3600)


# Raw input data with DMS and heights
data = [
    {"lat_d": 23, "lat_m": 48, "lat_s": 53.02690, "lon_d": 86, "lon_m": 26, "lon_s": 26.21279, "height": 193.149},
    {"lat_d": 23, "lat_m": 48, "lat_s": 52.30527, "lon_d": 86, "lon_m": 26, "lon_s": 25.55737, "height": 186.339},
    {"lat_d": 23, "lat_m": 48, "lat_s": 51.31074, "lon_d": 86, "lon_m": 26, "lon_s": 25.18608, "height": 188.559},
    {"lat_d": 23, "lat_m": 48, "lat_s": 49.07964, "lon_d": 86, "lon_m": 26, "lon_s": 27.84514, "height": 192.97},
    {"lat_d": 23, "lat_m": 48, "lat_s": 49.61347, "lon_d": 86, "lon_m": 26, "lon_s": 29.56767, "height": 190.347},
    {"lat_d": 23, "lat_m": 48, "lat_s": 46.62341, "lon_d": 86, "lon_m": 26, "lon_s": 31.30178, "height": 191.72},
    {"lat_d": 23, "lat_m": 48, "lat_s": 41.12877, "lon_d": 86, "lon_m": 26, "lon_s": 32.69507, "height": 190.228},
    {"lat_d": 23, "lat_m": 48, "lat_s": 43.19376, "lon_d": 86, "lon_m": 26, "lon_s": 36.55444, "height": 188.681},
    {"lat_d": 23, "lat_m": 48, "lat_s": 47.68798, "lon_d": 86, "lon_m": 26, "lon_s": 36.73159, "height": 188.334},
    {"lat_d": 23, "lat_m": 48, "lat_s": 53.47447, "lon_d": 86, "lon_m": 26, "lon_s": 36.57667, "height": 183.705},
]

# Parse data and convert DMS to Decimal
parsed_data = []
for entry in data:
    lat_decimal = dms_to_decimal(entry["lat_d"], entry["lat_m"], entry["lat_s"])
    lon_decimal = dms_to_decimal(entry["lon_d"], entry["lon_m"], entry["lon_s"])
    parsed_data.append({"latitude": lat_decimal, "longitude": lon_decimal, "height": entry["height"]})

df = pd.DataFrame(parsed_data)

# Apply Bowditch Rule correction
# Using a hypothetical scaling correction factor
scale_factor_lat = 0.999  # Scale factor for latitude correction
scale_factor_lon = 0.998  # Scale factor for longitude correction

df["corrected_latitude"] = df["latitude"] * scale_factor_lat
df["corrected_longitude"] = df["longitude"] * scale_factor_lon

# Compute errors
df["error_latitude"] = np.abs(df["latitude"] - df["corrected_latitude"])
df["error_longitude"] = np.abs(df["longitude"] - df["corrected_longitude"])

# Define polygon by adding boundary - close the loop
boundary_lat_raw = df["latitude"].tolist() + [df["latitude"].iloc[0]]  # Close the loop by adding the first point again
boundary_lon_raw = df["longitude"].tolist() + [df["longitude"].iloc[0]]

boundary_lat_corrected = df["corrected_latitude"].tolist() + [df["corrected_latitude"].iloc[0]]
boundary_lon_corrected = df["corrected_longitude"].tolist() + [df["corrected_longitude"].iloc[0]]


# Calculate area using the Shoelace formula
def calculate_area(lats, lons):
    """Shoelace formula to compute area"""
    n = len(lats)
    area = 0
    for i in range(n - 1):
        area += (lons[i] * lats[i + 1]) - (lons[i + 1] * lats[i])
    area = abs(area) / 2
    return area


# Calculate boundary length
def calculate_boundary_length(lats, lons):
    """Sum up the distances between consecutive points to compute the boundary length"""
    length = 0
    for i in range(len(lats) - 1):
        length += np.sqrt((lats[i + 1] - lats[i])**2 + (lons[i + 1] - lons[i])**2)
    return length


# Compute area and boundary lengths
area_raw = calculate_area(boundary_lat_raw, boundary_lon_raw)
area_corrected = calculate_area(boundary_lat_corrected, boundary_lon_corrected)

boundary_length_raw = calculate_boundary_length(boundary_lat_raw, boundary_lon_raw)
boundary_length_corrected = calculate_boundary_length(boundary_lat_corrected, boundary_lon_corrected)

# Plotting the raw and corrected data with boundary
plt.figure(figsize=(15, 7))

# Raw Data with polygon boundary
plt.subplot(1, 2, 1)
plt.scatter(df["longitude"], df["latitude"], c='blue', label="Raw Data")
plt.plot(boundary_lon_raw, boundary_lat_raw, 'r-', label="Boundary (Raw Data)")
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title("Raw GNNS Data with Boundary")
plt.legend()

# Corrected Data with polygon boundary
plt.subplot(1, 2, 2)
plt.scatter(df["corrected_longitude"], df["corrected_latitude"], c='green', label="Corrected Data")
plt.plot(boundary_lon_corrected, boundary_lat_corrected, 'r-', label="Boundary (Corrected Data)")
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title("Corrected GNNS Data with Boundary")
plt.legend()

plt.tight_layout()
plt.show()

# Output the tables for raw and corrected data
print("\nTable of Raw Coordinates:")
print(df[["latitude", "longitude", "height"]])

print("\nTable of Corrected Coordinates:")
print(df[["corrected_latitude", "corrected_longitude", "height"]])

# Output calculated errors and areas
print("\nErrors:")
print(f"Total Latitude Errors: {df['error_latitude'].sum()}")
print(f"Total Longitude Errors: {df['error_longitude'].sum()}")

print("\nCalculated Areas:")
print(f"Raw GNNS Data Area: {area_raw} square degrees")
print(f"Corrected GNNS Data Area: {area_corrected} square degrees")

print("\nBoundary Lengths:")
print(f"Raw GNNS Data Boundary Length: {boundary_length_raw} degrees")
print(f"Corrected GNNS Data Boundary Length: {boundary_length_corrected} degrees")
```
